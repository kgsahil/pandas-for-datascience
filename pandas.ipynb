 Be careful while modifying data, always reassign values via .loc\[\] or .at[]
 When filter is applied to the data without using 'loc' or 'at' it returns the copy of the dataframe statisfied to filter
 eg: poeple_df\[(some kind of filter)\] will gives copy, modification of the data will fail.
 the link in warning explains the diffrence.

```python
filt = people_df['first'] == 'sahil'
people_df[filt]['last'] = 'Gupta'
```

To modify data successfully use .loc or .at with filter.

```python
people_df.loc[filt,['last']] = 'Gupta'
people_df
```

### four methods for accessing and updating
### 1) apply
### 2) applymap
### 3) map
### 4) replace

1) apply

```python
people_df['email'].apply(len)
```

```python
def update_email(email):
    return email.upper()

people_df['email'].apply(update_email) 
#this does'nt change the original values to  do it just with the update columns
```

```python
people_df['email'].apply(lambda x: x.lower())
```

2) applymap, only used for Dataframe for every individual values

 3) map function takes key:value pairs which replace each occurence of the key with the given value and makes every "not mentioned" keys to NaN
