 Be careful while modifying data, always reassign values via .loc\[\] or .at[]
 When filter is applied to the data without using 'loc' or 'at' it returns the copy of the dataframe statisfied to filter
 eg: poeple_df\[(some kind of filter)\] will gives copy, modification of the data will fail.
 the link in warning explains the diffrence.

```python
filt = people_df['first'] == 'sahil'
people_df[filt]['last'] = 'Gupta'
```

To modify data successfully use .loc or .at with filter.

```python
people_df.loc[filt,['last']] = 'Gupta'
people_df
```

# Part 8: Casting dtype and handling missing values

```python
people_df = people_df.append({'first':'juned','last':'khan','full_name':'juned khan'},ignore_index=True)
people_df
```

```python
#remove missing data
people_df.dropna()
```

```python
#dropna with default argument
people_df.dropna(axis='index', how='all') #only delete rows when all the data is missing
people_df.dropna(axis='columns', how='all') #only delete columns when all the data is missing
people_df.dropna(axis='columns', how='any') #only delete columns when any one of them the data is missing
```

```python
people_df.dropna(axis='index', how='all',subset=['email']) #only delete specific rows when any one of them the data is missing
```
